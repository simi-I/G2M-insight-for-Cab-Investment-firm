## Data Merging 

```python
Custs_trans_data = pd.merge(customer_data, transaction_data, how = 'outer', on = 'Customer ID')
```

```python
cab_city_data = pd.merge(cab_data, city_data, how='outer', on = 'City')
```

```python
all_data_merge = pd.merge(Custs_trans_data, cab_city_data, how = 'outer', on = 'Transaction ID')
```

## Feature Engineering

```python
# Transform from days into date

def dateTransformation(dateInt):
    date = datetime.strftime(datetime(1899,12,29) + timedelta(days=dateInt),'%d-%m-%Y')
    date = datetime.strptime(date, "%d-%m-%Y")
    return date
```

```python
all_data_merge['Date of Travel']=all_data_merge['Date of Travel'].apply(dateTransformation) 
```

```python
print(all_data_merge['Date of Travel'].min())
print(all_data_merge['Date of Travel'].max())
```

```python
# month of date
all_data_merge['Month'] = pd.DatetimeIndex(all_data_merge['Date of Travel']).month

# day of datetime
all_data_merge['Day'] = pd.DatetimeIndex(all_data_merge['Date of Travel']).day

# year of datetime
all_data_merge['Year'] = pd.DatetimeIndex(all_data_merge['Date of Travel']).year
```

```python
#Calculate Profit
all_data_merge['Profit'] = all_data_merge['Price Charged'] - all_data_merge['Cost of Trip']

#Calculate Profit per KM
all_data_merge['Profit_per_KM'] = all_data_merge['Profit']/all_data_merge['KM Travelled']

#Calculate Cost per KM
all_data_merge['Cost_per_KM'] = all_data_merge['Cost of Trip']/all_data_merge['KM Travelled']
```

```python
# dataframe with timedate as index
all_data_merge_dt = all_data_merge.set_index('Date of Travel')
```
